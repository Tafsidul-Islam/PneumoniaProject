```python
# ============================================
# PNEUMONIA DETECTION USING RANDOM FOREST
# Complete Pipeline for Kaggle - Fixed LBP
# ============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2
from glob import glob
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

from skimage.feature import hog, local_binary_pattern
from skimage.transform import resize
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (classification_report, confusion_matrix, 
                           accuracy_score, precision_score, recall_score, 
                           f1_score, roc_auc_score, roc_curve)
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.utils import class_weight
from sklearn.model_selection import train_test_split

np.random.seed(42)

# ============================================
# 1. FIND THE ACTUAL DATASET FOLDER
# ============================================

BASE_PATH = '/kaggle/input/datasets/mtishakil'

print("="*60)
print("DATASET EXPLORATION")
print("="*60)

current_path = BASE_PATH
print(f"\nExploring: {current_path}")
print(f"Contents: {os.listdir(current_path)}")

for item in os.listdir(current_path):
    item_path = os.path.join(current_path, item)
    if os.path.isdir(item_path):
        print(f"\n📁 {item}/")
        sub_items = os.listdir(item_path)
        print(f"   Contents: {sub_items}")
        
        sub_items_lower = [s.lower() for s in sub_items]
        if any('train' in s for s in sub_items_lower):
            BASE_PATH = item_path
            print(f"   ✅ Found dataset folders here!")
            break
        
        for sub_item in sub_items:
            sub_path = os.path.join(item_path, sub_item)
            if os.path.isdir(sub_path):
                print(f"   📁 {sub_item}/")
                deep_items = os.listdir(sub_path)
                print(f"      Contents: {deep_items}")
                deep_items_lower = [d.lower() for d in deep_items]
                if any('train' in d for d in deep_items_lower):
                    BASE_PATH = sub_path
                    print(f"      ✅ Found dataset folders here!")
                    break

print(f"\n✅ Final BASE_PATH: {BASE_PATH}")
print(f"Contents: {os.listdir(BASE_PATH)}")

# ============================================
# 2. DATA LOADING FUNCTIONS
# ============================================

def load_image(image_path, target_size=(128, 128)):
    """Load and preprocess a single image"""
    try:
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            return None
        img = resize(img, target_size, anti_aliasing=True)
        img = img.astype(np.float32) / 255.0
        return img
    except:
        return None

def load_dataset_from_folder(data_dir, target_size=(128, 128), dataset_name="Dataset"):
    """Load dataset from a folder containing 'normal' and 'pneumonia' subfolders"""
    if data_dir is None or not os.path.exists(data_dir):
        print(f"⚠️  Directory not found: {data_dir}")
        return np.array([]), np.array([])
    
    images = []
    labels = []
    normal_count = 0
    pneumonia_count = 0
    
    print(f"\n📊 Loading {dataset_name} from: {data_dir}")
    
    for class_folder in os.listdir(data_dir):
        class_path = os.path.join(data_dir, class_folder)
        if os.path.isdir(class_path):
            folder_lower = class_folder.lower()
            
            if 'normal' in folder_lower:
                label = 0
                class_name = "Normal"
            elif any(term in folder_lower for term in ['pneumonia', 'bacteria', 'virus']):
                label = 1
                class_name = "Pneumonia"
            else:
                continue
            
            image_paths = []
            for file in os.listdir(class_path):
                if file.lower().endswith(('.png', '.jpg', '.jpeg', '.jfif', '.tiff', '.bmp')):
                    image_paths.append(os.path.join(class_path, file))
            
            print(f"   {class_name}: {len(image_paths)} images")
            
            for path in tqdm(image_paths, desc=f"   Loading {class_name}"):
                img = load_image(path, target_size)
                if img is not None:
                    images.append(img)
                    labels.append(label)
                    if label == 0:
                        normal_count += 1
                    else:
                        pneumonia_count += 1
    
    print(f"   ✅ Loaded {len(images)} images ({normal_count} normal, {pneumonia_count} pneumonia)")
    return np.array(images), np.array(labels)

# ============================================
# 3. FEATURE EXTRACTION (FIXED)
# ============================================

def extract_hog_features(images, desc="Extracting HOG"):
    """Extract Histogram of Oriented Gradients features"""
    if len(images) == 0:
        return np.array([])
    
    hog_features = []
    for img in tqdm(images, desc=desc):
        fd, _ = hog(img, 
                    orientations=9, 
                    pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), 
                    visualize=True,
                    channel_axis=None)
        hog_features.append(fd)
    return np.array(hog_features)

def extract_lbp_features(images, desc="Extracting LBP"):
    """Extract Local Binary Pattern features with fixed histogram size"""
    if len(images) == 0:
        return np.array([])
    
    # Fixed parameters for LBP
    radius = 3
    n_points = 24
    n_bins = 26  # Fixed number of bins for 'uniform' method
    
    lbp_features = []
    for img in tqdm(images, desc=desc):
        img_uint8 = (img * 255).astype(np.uint8)
        lbp = local_binary_pattern(img_uint8, n_points, radius, method='uniform')
        
        # Use fixed number of bins
        hist, _ = np.histogram(lbp, bins=n_bins, range=(0, n_bins), density=True)
        lbp_features.append(hist)
    
    return np.array(lbp_features)

def extract_statistical_features(images, desc="Extracting statistics"):
    """Extract statistical features"""
    if len(images) == 0:
        return np.array([])
    
    stat_features = []
    for img in tqdm(images, desc=desc):
        features = [
            np.mean(img),
            np.std(img),
            np.median(img),
            np.min(img),
            np.max(img),
            np.percentile(img, 25),
            np.percentile(img, 75),
            float(pd.Series(img.flatten()).skew()),
            float(pd.Series(img.flatten()).kurtosis())
        ]
        hist, _ = np.histogram(img.flatten(), bins=50, density=True)
        features.append(np.sum(hist ** 2))  # Energy
        features.append(-np.sum(hist * np.log2(hist + 1e-10)))  # Entropy
        stat_features.append(features)
    return np.array(stat_features)

def extract_all_features(images, desc="Processing"):
    """Extract all features combined"""
    if len(images) == 0:
        return np.array([])
    
    print(f"\n   {desc}:")
    hog_feat = extract_hog_features(images, "     • HOG features")
    lbp_feat = extract_lbp_features(images, "     • LBP features")
    stat_feat = extract_statistical_features(images, "     • Statistical features")
    
    if len(hog_feat) > 0 and len(lbp_feat) > 0 and len(stat_feat) > 0:
        print(f"     HOG: {hog_feat.shape[1]} features")
        print(f"     LBP: {lbp_feat.shape[1]} features")
        print(f"     Statistical: {stat_feat.shape[1]} features")
        
        # Ensure all arrays are 2D before stacking
        combined = np.hstack([hog_feat, lbp_feat, stat_feat])
        print(f"     Total: {combined.shape[1]} features")
        return combined
    return np.array([])

# ============================================
# 4. VISUALIZATION
# ============================================

def plot_sample_images(images, labels, num_samples=5):
    """Plot sample images"""
    if len(images) == 0:
        return
    
    normal_idx = np.where(labels == 0)[0]
    pneumonia_idx = np.where(labels == 1)[0]
    
    num_samples = min(num_samples, len(normal_idx), len(pneumonia_idx))
    
    fig, axes = plt.subplots(2, num_samples, figsize=(15, 6))
    
    if num_samples == 1:
        axes = axes.reshape(2, 1)
    
    for i in range(num_samples):
        axes[0, i].imshow(images[normal_idx[i]], cmap='gray')
        axes[0, i].set_title(f'Normal')
        axes[0, i].axis('off')
    
    for i in range(num_samples):
        axes[1, i].imshow(images[pneumonia_idx[i]], cmap='gray')
        axes[1, i].set_title(f'Pneumonia')
        axes[1, i].axis('off')
    
    plt.suptitle('Sample Chest X-Ray Images', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()

def plot_results(y_true, y_pred, y_prob):
    """Plot confusion matrix and ROC curve"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax1,
                xticklabels=['Normal', 'Pneumonia'],
                yticklabels=['Normal', 'Pneumonia'])
    ax1.set_title('Confusion Matrix', fontsize=14, fontweight='bold')
    ax1.set_ylabel('True Label')
    ax1.set_xlabel('Predicted Label')
    
    fpr, tpr, _ = roc_curve(y_true, y_prob[:, 1])
    auc = roc_auc_score(y_true, y_prob[:, 1])
    
    ax2.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC (AUC = {auc:.3f})')
    ax2.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    ax2.set_xlim([0.0, 1.0])
    ax2.set_ylim([0.0, 1.05])
    ax2.set_xlabel('False Positive Rate')
    ax2.set_ylabel('True Positive Rate')
    ax2.set_title('ROC Curve', fontsize=14, fontweight='bold')
    ax2.legend(loc="lower right")
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# ============================================
# 5. MAIN PIPELINE
# ============================================

def main():
    IMAGE_SIZE = (128, 128)
    USE_PCA = True
    PCA_COMPONENTS = 200
    
    print("\n" + "="*60)
    print("PNEUMONIA DETECTION USING RANDOM FOREST")
    print("="*60)
    
    # Find train/val/test folders
    train_dir = None
    val_dir = None
    test_dir = None
    
    print(f"\nLooking for train/val/test folders in: {BASE_PATH}")
    
    for item in os.listdir(BASE_PATH):
        item_path = os.path.join(BASE_PATH, item)
        if os.path.isdir(item_path):
            item_lower = item.lower()
            print(f"   Found folder: {item}")
            if 'train' in item_lower:
                train_dir = item_path
            elif 'val' in item_lower or 'valid' in item_lower:
                val_dir = item_path
            elif 'test' in item_lower:
                test_dir = item_path
    
    print(f"\n📁 Using directories:")
    print(f"   Train: {train_dir}")
    print(f"   Validation: {val_dir}")
    print(f"   Test: {test_dir}")
    
    if train_dir is None:
        base_contents = [d.lower() for d in os.listdir(BASE_PATH)]
        if 'normal' in base_contents or 'pneumonia' in base_contents:
            print("\n   Using BASE_PATH as the dataset directory")
            train_dir = BASE_PATH
    
    # ============================================
    # Load Data
    # ============================================
    print("\n" + "="*60)
    print("STEP 1: LOADING DATA")
    print("="*60)
    
    X_train, y_train = load_dataset_from_folder(train_dir, IMAGE_SIZE, "Training Set")
    X_val, y_val = load_dataset_from_folder(val_dir, IMAGE_SIZE, "Validation Set") if val_dir else (np.array([]), np.array([]))
    X_test, y_test = load_dataset_from_folder(test_dir, IMAGE_SIZE, "Test Set") if test_dir else (np.array([]), np.array([]))
    
    if len(X_val) == 0 and len(X_test) == 0 and len(X_train) > 0:
        print("\n   No validation/test sets found. Splitting training data...")
        X_train, X_test, y_train, y_test = train_test_split(
            X_train, y_train, test_size=0.15, stratify=y_train, random_state=42
        )
        X_train, X_val, y_train, y_val = train_test_split(
            X_train, y_train, test_size=0.176, stratify=y_train, random_state=42
        )
    elif len(X_val) == 0 and len(X_train) > 0:
        X_train, X_val, y_train, y_val = train_test_split(
            X_train, y_train, test_size=0.15, stratify=y_train, random_state=42
        )
    elif len(X_test) == 0 and len(X_train) > 0:
        X_train, X_test, y_train, y_test = train_test_split(
            X_train, y_train, test_size=0.15, stratify=y_train, random_state=42
        )
    
    print(f"\n📊 Final Dataset Summary:")
    print(f"   Training:   {len(X_train):5d} images (Normal: {np.sum(y_train==0)}, Pneumonia: {np.sum(y_train==1)})")
    print(f"   Validation: {len(X_val):5d} images (Normal: {np.sum(y_val==0)}, Pneumonia: {np.sum(y_val==1)})")
    print(f"   Test:       {len(X_test):5d} images (Normal: {np.sum(y_test==0)}, Pneumonia: {np.sum(y_test==1)})")
    
    if len(X_train) == 0:
        print("\n❌ ERROR: No training data loaded!")
        return
    
    plot_sample_images(X_train, y_train)
    
    # ============================================
    # Feature Extraction
    # ============================================
    print("\n" + "="*60)
    print("STEP 2: FEATURE EXTRACTION")
    print("="*60)
    
    X_train_feat = extract_all_features(X_train, "Training features")
    X_val_feat = extract_all_features(X_val, "Validation features")
    X_test_feat = extract_all_features(X_test, "Test features")
    
    # ============================================
    # Feature Scaling
    # ============================================
    print("\n" + "="*60)
    print("STEP 3: FEATURE SCALING")
    print("="*60)
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_feat)
    X_val_scaled = scaler.transform(X_val_feat)
    X_test_scaled = scaler.transform(X_test_feat)
    print("   ✅ Features scaled")
    
    # ============================================
    # PCA
    # ============================================
    if USE_PCA:
        print("\n" + "="*60)
        print("STEP 4: DIMENSIONALITY REDUCTION (PCA)")
        print("="*60)
        
        n_components = min(PCA_COMPONENTS, X_train_scaled.shape[1])
        pca = PCA(n_components=n_components)
        X_train_final = pca.fit_transform(X_train_scaled)
        X_val_final = pca.transform(X_val_scaled)
        X_test_final = pca.transform(X_test_scaled)
        
        print(f"   Original features: {X_train_scaled.shape[1]}")
        print(f"   Reduced features:  {X_train_final.shape[1]}")
        print(f"   Explained variance: {pca.explained_variance_ratio_.sum():.3f} ({pca.explained_variance_ratio_.sum()*100:.1f}%)")
    else:
        X_train_final, X_val_final, X_test_final = X_train_scaled, X_val_scaled, X_test_scaled
    
    # ============================================
    # Train Model
    # ============================================
    print("\n" + "="*60)
    print("STEP 5: TRAINING RANDOM FOREST")
    print("="*60)
    
    class_weights = class_weight.compute_class_weight(
        'balanced',
        classes=np.unique(y_train),
        y=y_train
    )
    class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
    print(f"   Class weights: Normal={class_weight_dict[0]:.3f}, Pneumonia={class_weight_dict[1]:.3f}")
    
    rf_model = RandomForestClassifier(
        n_estimators=200,
        max_depth=20,
        min_samples_split=5,
        min_samples_leaf=2,
        max_features='sqrt',
        class_weight=class_weight_dict,
        random_state=42,
        n_jobs=-1,
        verbose=1
    )
    
    print("   Training model...")
    rf_model.fit(X_train_final, y_train)
    print("   ✅ Training completed!")
    
    # ============================================
    # Evaluation
    # ============================================
    print("\n" + "="*60)
    print("STEP 6: MODEL EVALUATION")
    print("="*60)
    
    y_train_pred = rf_model.predict(X_train_final)
    train_acc = accuracy_score(y_train, y_train_pred)
    
    y_val_pred = rf_model.predict(X_val_final)
    y_val_prob = rf_model.predict_proba(X_val_final)
    
    y_test_pred = rf_model.predict(X_test_final)
    y_test_prob = rf_model.predict_proba(X_test_final)
    
    print(f"\n📊 Performance Summary:")
    print(f"   {'Metric':<20} {'Training':<12} {'Validation':<12} {'Test':<12}")
    print(f"   {'-'*20} {'-'*12} {'-'*12} {'-'*12}")
    print(f"   {'Accuracy':<20} {train_acc:<12.4f} {accuracy_score(y_val, y_val_pred):<12.4f} {accuracy_score(y_test, y_test_pred):<12.4f}")
    print(f"   {'Precision':<20} {'':<12} {precision_score(y_val, y_val_pred):<12.4f} {precision_score(y_test, y_test_pred):<12.4f}")
    print(f"   {'Recall':<20} {'':<12} {recall_score(y_val, y_val_pred):<12.4f} {recall_score(y_test, y_test_pred):<12.4f}")
    print(f"   {'F1-Score':<20} {'':<12} {f1_score(y_val, y_val_pred):<12.4f} {f1_score(y_test, y_test_pred):<12.4f}")
    print(f"   {'AUC-ROC':<20} {'':<12} {roc_auc_score(y_val, y_val_prob[:, 1]):<12.4f} {roc_auc_score(y_test, y_test_prob[:, 1]):<12.4f}")
    
    print(f"\n📋 Classification Report (Test Set):")
    print(classification_report(y_test, y_test_pred, target_names=['Normal', 'Pneumonia']))
    
    cm = confusion_matrix(y_test, y_test_pred)
    tn, fp, fn, tp = cm.ravel()
    
    print(f"🏥 Medical Relevance:")
    print(f"   True Negatives (Correctly identified Normal): {tn}")
    print(f"   False Positives (Normal → Pneumonia):         {fp}")
    print(f"   False Negatives (Missed Pneumonia):            {fn} {'⚠️' if fn > 0 else '✅'}")
    print(f"   True Positives (Correctly identified Pneumonia): {tp}")
    print(f"   Sensitivity/Recall: {tp/(tp+fn):.4f}")
    print(f"   Specificity:        {tn/(tn+fp):.4f}")
    
    plot_results(y_test, y_test_pred, y_test_prob)
    
    importances = rf_model.feature_importances_
    top_20_idx = np.argsort(importances)[-20:]
    
    plt.figure(figsize=(10, 6))
    plt.barh(range(20), importances[top_20_idx])
    plt.yticks(range(20), [f'Feature {i}' for i in top_20_idx])
    plt.xlabel('Importance')
    plt.title('Top 20 Feature Importances')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
    print("\n" + "="*60)
    print("✅ PIPELINE COMPLETED SUCCESSFULLY!")
    print("="*60)
    
    return rf_model

# ============================================
# RUN THE PIPELINE
# ============================================
if __name__ == "__main__":
    model = main()
```
